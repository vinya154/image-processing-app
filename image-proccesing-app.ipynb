```python
import gradio as gr
```

```python
def IMA(img,ch):
  if(ch=='BLUR'):
    img=img.filter(ImageFilter.BLUR)
  elif(ch=='DETAIL'):
    img=img.filter(ImageFilter.DETAIL)
  elif(ch=='GRAY'):
    img=img.convert('L')
  return img
```

```python
iface=gr.Interface(
    fn=IMA,
    inputs=[gr.Image(type='pil',label='Upload Image'),gr.Radio(choices=['BLUR','DETAIL','GRAY'])],
    outputs=[gr.Image()]
)
iface.launch()
```
